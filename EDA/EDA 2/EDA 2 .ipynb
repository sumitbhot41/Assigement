```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the flight price dataset
flight_data = pd.read_csv(r"C:\Users\sumit bhot\Downloads\flight price (7)\economy.csv")

# Q1: Examine the dimensions of the dataset
print("Dimensions of the dataset:", flight_data.shape)  # This will print the number of rows and columns

```

```python
# Q2: Distribution of flight prices
plt.figure(figsize=(10, 6))
sns.histplot(flight_data['price'], bins=20, kde=True)
plt.title('Distribution of Flight Prices')
plt.xlabel('price')
plt.ylabel('Frequency')
plt.show()
```

```python
print(flight_data.columns)
```

```python

# Q3: Range of prices
price_min = flight_data['price'].min()
price_max = flight_data['price'].max()
print("Minimum Price:", price_min)
print("Maximum Price:", price_max)


```

```python
# Q4: Price of flights by airline
plt.figure(figsize=(12, 6))
sns.boxplot(x='airline', y='price', data=flight_data)
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('price')
plt.xticks(rotation=45)
plt.show()

```

```python
# Q5: Identifying outliers using a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=flight_data['price'])
plt.title('Boxplot of Flight Prices')
plt.xlabel('Price')
plt.show()
# Here, you can visually identify any potential outliers beyond the whiskers of the boxplot.

```
