```python
import seaborn as sns
import plotly.express as px

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Plot scatter plot using Plotly Express
fig = px.scatter(titanic_data, x="age", y="fare", title="Titanic: Age vs Fare")
fig.show()
```

```python
tips_data = px.data.tips()
fig2 = px.box(tips_data, x="day", y="total_bill", title="Total Bill Distribution by Day")
fig2.show()
```

```python
fig3 = px.histogram(tips_data, x="sex", y="total_bill", color="smoker", 
                    facet_col="day", facet_row="time", 
                    pattern_shape="smoker", color_discrete_sequence=px.colors.qualitative.Set1,
                    title="Total Bill Distribution by Sex, Smoker, and Day")
fig3.show()
```

```python
iris_data = px.data.iris()
fig4 = px.scatter_matrix(iris_data, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
                         color="species", title="Scatter Matrix Plot of Iris Dataset")
fig4.show()
```

```python
fig5 = px.histogram(tips_data, x="total_bill", marginal="rug", title="Distribution of Total Bill")
fig5.show()
```
