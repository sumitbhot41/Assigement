```python
def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    # Using sample mean as point estimate for population mean
    population_mean = sample_mean
    # Adjusting for smaller sample sizes using t-distribution
    population_mean -= (1.96 * sample_std_dev / (sample_size ** 0.5))
    return population_mean

# Example usage:
sample_mean = 75
sample_std_dev = 8
sample_size = 40
population_mean_estimate = estimate_population_mean(sample_mean, sample_std_dev, sample_size)
print("Estimated Population Mean:", population_mean_estimate)
```

```python
Q5: Python script for hypothesis test on difference between two population means:

```

```python
import scipy.stats

def two_sample_t_test(sample1, sample2):
    # Perform two-sample t-test
    t_statistic, p_value = scipy.stats.ttest_ind(sample1, sample2)
    return t_statistic, p_value

# Example usage:
sample1 = [83, 86, 88, 82, 85]
sample2 = [78, 80, 85, 81, 79]
t_statistic, p_value = two_sample_t_test(sample1, sample2)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```
