```python
Types of plots in Seaborn and their uses:

Scatter Plot: Useful for visualizing the relationship between two continuous variables.
Line Plot: Effective for showing trends over time or any ordered dimension.
Bar Plot: Suitable for comparing categorical variables.
Histogram: Ideal for understanding the distribution of a single variable.
Box Plot: Helpful in displaying the distribution of data based on the quartiles
```

```python
Load the "fmri" dataset and plot a line plot:
```

```python
import seaborn as sns

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
```

```python
Load the "titanic" dataset and plot two box plots:
```

```python
# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots
sns.boxplot(x="pclass", y="age", data=titanic_data)
sns.boxplot(x="pclass", y="fare", data=titanic_data)
```

```python
diamonds_data = sns.load_dataset("diamonds")
sns.histplot(data=diamonds_data, x="price", hue="cut", bins=30)
```

```python
iris_data = sns.load_dataset("iris")
sns.pairplot(iris_data, hue="species")
```

```python
flights_data = sns.load_dataset("flights")
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, cmap="YlGnBu")
```
