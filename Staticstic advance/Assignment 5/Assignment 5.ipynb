```python
import numpy as np
import scipy.stats as stats

# Q1: Calculate the 95% confidence interval
mean = 50
std_dev = 5
n = 1  # sample size
confidence_level = 0.95

margin_of_error = stats.norm.ppf(1 - (1 - confidence_level) / 2) * (std_dev / np.sqrt(n))
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

print("95% Confidence Interval:", confidence_interval)
# Interpretation: We are 95% confident that the true population mean lies within the interval (47.5, 52.5).
```

```python
# Q2: Chi-square goodness of fit test
observed = [20, 20, 20, 10, 10, 20]
expected = [0.2, 0.2, 0.2, 0.1, 0.1, 0.2]  # Expected proportions
chi2, p = stats.chisquare(observed, f_exp=np.array(expected) * sum(observed))

print("Chi-square statistic:", chi2)
print("p-value:", p)
# Interpretation: Based on the p-value (p < 0.05), we reject the null hypothesis that the observed distribution
# matches the expected distribution.
```

```python
# Q3: Chi-square test for contingency table
observed = np.array([[20, 15], [10, 25], [15, 20]])
chi2, p, _, _ = stats.chi2_contingency(observed)

print("Chi-square statistic:", chi2)
print("p-value:", p)
# Interpretation: The p-value is greater than 0.05, indicating that there is no significant association between
# smoking status and lung cancer diagnosis.
```

```python
# Q4: Calculate the 95% confidence interval for proportion
n = 500
smokers = 60
p_hat = smokers / n
margin_of_error = stats.norm.ppf(1 - (1 - 0.95) / 2) * np.sqrt((p_hat * (1 - p_hat)) / n)
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)

print("95% Confidence Interval for proportion of smokers:", confidence_interval)
```

```python
# Q6: Plot chi-square distribution
import matplotlib.pyplot as plt

df = 10  # degrees of freedom
x = np.linspace(0, 30, 1000)
plt.plot(x, stats.chi2.pdf(x, df), 'b-', label='Chi-square distribution')
plt.fill_between(x, 0, stats.chi2.pdf(x, df), where=(x >= 15), color="red", alpha=0.3)
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability density')
plt.title('Chi-square Distribution (df=10)')
plt.legend()
plt.show()
```

```python
# Q7: Calculate the 99% confidence interval
n = 1000
coke_pref = 520
p_hat = coke_pref / n
margin_of_error = stats.norm.ppf(1 - (1 - 0.99) / 2) * np.sqrt((p_hat * (1 - p_hat)) / n)
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)

print("99% Confidence Interval for proportion of Coke preference:", confidence_interval)
```
