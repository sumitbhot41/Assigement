```python
ZeroDivisionError: Raised when division or modulo by zero occurs.
```

```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Error:", e)
```

```python
OverflowError: Raised when the result of an arithmetic operation is too large to be represented.
```

```python
try:
    result = 10 ** 10000
except OverflowError as e:
    print("Error:", e)
```

```python
Q5. Explain ImportError. What is ModuleNotFoundError?
```

```python
A5. ImportError is a base class for exceptions raised when a module could not be imported. ModuleNotFoundError is a subclass of ImportError and is raised when a module could not be found.

ImportError: Raised when a module is found but cannot be imported due to various reasons such as syntax errors in the module or missing dependencies.
ModuleNotFoundError: Raised when a module could not be found during import.
```
